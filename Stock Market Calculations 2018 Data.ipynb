```r
apple_df<- read.csv("aapl-1.csv")
amazon_df<- read.csv("amzn-1.csv")
google_df<- read.csv("goog-1.csv")
ibm_df<- read.csv("ibm-1.csv")
meta_df<- read.csv("meta-1.csv")
msft_df<- read.csv("msft-1.csv")
```

```r
#Checking variable names and types
```

```r
str(apple_df)
```

```r
#Code to calculate APPLE close ratio, adjusted close ratio, difference, and divided
```

```r
apple_df2 <- apple_df %>%
  mutate(
    Close_Ratio = lead(Close, n = 1) / Close,
    Adjusted_Close_Ratio = lead(Adj.Close, n = 1) / Adj.Close,
    Difference = Adjusted_Close_Ratio - Close_Ratio,
    Dividend = Difference * Close
    )
apple_df2
```

```r
#Code for APPLE table of date and dividend only.
```

```r
select(apple_df2, Date, Dividend)
```
